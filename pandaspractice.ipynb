```python
# loc for lebal base selection
# iloc for index based selection
```

```python
#LOC .. i want PG-13 ratings films only 
data.loc[data.rating=='PG-13']
```

```python
#Loc .. i want PG-13 and 2011 release year films
data.loc[(data.rating=='PG-13') & (data["release year"] == 2011)]
```

```python
#Loc .. i want PG-13 or 2011 release year films
data.loc[(data.rating=='PG-13') | (data["release year"] == 2011)]
```

```python
#loc .. i want check films based 90 or with above user user rating score 
data.loc[data["user rating score"]>=90]
```

```python
#in core python first columns passes then rows
# but in pandas first row passes then columns
```

```python
# i want selected columns along with desired column
data.loc[data.rating=='PG-13',['title']]
#it will show me PG-13 columns but i olny wanted titles not user score all other stuff 
```

```python
# for example i want ratingLevel along with tile also
data.loc[data.rating=='PG-13',['title','ratingLevel']]
```
